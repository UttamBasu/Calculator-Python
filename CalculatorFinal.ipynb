```python
from tkinter import*
import random
import time

root = Tk()
root.geometry("375x650+0+0")
root.title("Calculator for You")

text_Input = StringVar()
operator = ""

Tops = Frame(root, width = 400,height = 50, bg ="#5c3854", relief=SUNKEN)
Tops.pack(side=TOP)


f2 = Frame(root, width = 400, height = 550, bg ="Powder blue", relief=SUNKEN)
f2.pack(side=LEFT)
#============================Time============================================
localtime=time.asctime(time.localtime(time.time()))
#=================================Info==========================================
lblInfo = Label(Tops, font=('arial',28,'bold'),text=" Calculator for You ", fg="Steel Blue",bd=0, anchor='w')
lblInfo.grid(row=0, column=0)
   
lblInfo = Label(Tops, font=('arial',8,'bold'),text="                                                                        -By Uttam Basu              ", fg="#408c8c", bd=1, anchor='w')
lblInfo.grid(row=1, column=0)
 
lblInfo = Label(Tops, font=('arial',12,'bold'),text="   Date & Time-      "+localtime,fg="#ff2999", bd=5, anchor='w')
lblInfo.grid(row=2, column=0)
#================================Calculator========================================
def btnClick(numbers):
    global operator
    operator=operator + str(numbers)
    text_Input.set(operator)
    
def btnClearDisplay():
    global operator
    operator=""
    text_Input.set("")
    
def btnEqualsInput():
    global operator
    sumup=str(eval(operator))
    text_Input.set(sumup)
    operator=""

txtDisplay = Entry(f2, font=('arial',20,'bold'), textvariable=text_Input, bd=30, insertwidth=4, bg="#8c947d", justify='right')
txtDisplay.grid(columnspan=4)

#=================================Row1===================================================
btn7=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="7", bg= "#CC99FF", command = lambda: btnClick(7)).grid(row=2,column=0)
btn8=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="8", bg= "#CC99FF", command = lambda: btnClick(8)).grid(row=2,column=1)
btn9=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="9", bg= "#CC99FF", command = lambda: btnClick(9)).grid(row=2,column=2)
Addition=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="+", bg= "#00CC99", command = lambda: btnClick("+")).grid(row=2,column=3)

#==================================Row2====================================================
btn4=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="4", bg= "#CC66FF", command = lambda: btnClick(4)).grid(row=3,column=0)
btn5=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="5", bg= "#CC66FF", command = lambda: btnClick(5)).grid(row=3,column=1)
btn6=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="6", bg= "#CC66FF", command = lambda: btnClick(6)).grid(row=3,column=2)
Subtraction=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="-", bg= "#009999", command = lambda: btnClick("-")).grid(row=3,column=3)

#==================================Row3====================================================
btn1=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="1", bg= "#CC33FF", command = lambda: btnClick(1)).grid(row=4,column=0)
btn2=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="2", bg= "#CC33FF", command = lambda: btnClick(2)).grid(row=4,column=1)
btn3=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="3", bg= "#CC33FF", command = lambda: btnClick(3)).grid(row=4,column=2)
Multiply=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="*", bg= "#006699", command = lambda: btnClick("*")).grid(row=4,column=3)

#==================================Row4====================================================
btn0=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="0", bg= "#521499", command = lambda: btnClick(0)).grid(row=5,column=0)
btnClear=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="C", bg= "Red", command=btnClearDisplay).grid(row=5,column=1)
btnEquals=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="=", bg= "Green", command=btnEqualsInput).grid(row=5,column=2)
Division=Button(f2,padx=16,pady=16,bd=8, fg="black", font=('arial',20,'bold'), text="/", bg= "#336699", command = lambda: btnClick("/")).grid(row=5,column=3)
root.mainloop()
```
